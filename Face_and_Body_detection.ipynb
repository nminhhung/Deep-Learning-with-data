## Face and Body detection

### Face & glasses detection

```python
import cv2
import matplotlib.pyplot as plt
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
glasses_cascade = cv2.CascadeClassifier('haarcascade_eye_tree_eyeglasses_1.xml')
```

```python
image = cv2.imread('glasses_1.jpg')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
```

```python
for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = image[y:y+h, x:x+w]
```

```python
glasses = glasses_cascade.detectMultiScale(roi_gray)
```

```python
for (ex,ey,ew,eh) in glasses:
    cx = ex+ew//2
    cy = ey+eh//2
    cr  = max(ew,eh)//2
    ## set offset, repeat enlarger ROI
    dr = 4
    i = 3
    r = cr+i*dr
    cv2.rectangle(roi_color, (cx-r, cy-r), (cx+r, cy+r), (0,255,0), 1) 
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

### Body detection

```python
body_classifier = cv2.CascadeClassifier('haarcascade_fullbody.xml')
```

```python
img = cv2.imread('body.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
bodies = body_classifier.detectMultiScale(gray, 1.25, 5)

for (xb, yb, wb, hb) in bodies:
    cv2.rectangle(img, (xb, yb), (xb+wb, yb+hb), (0, 255, 255), 2)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
